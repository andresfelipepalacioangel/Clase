```python
# prompt: me ayudas a graficar la funcion seno

import matplotlib.pyplot as plt
import numpy as np

# Generar datos para la función seno
x = np.linspace(0, 2 * np.pi, 100)  # 100 puntos entre 0 y 2*pi
y = np.sin(x)

# Crear la gráfica
plt.plot(x, y)

# Agregar etiquetas y título
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.title("Gráfica de la función seno")

# Mostrar la gráfica
plt.grid(True)
plt.show()
```

```python
# prompt: cambiar el color de la linea a rojo

import matplotlib.pyplot as plt
import numpy as np


# Generar datos para la función seno
x = np.linspace(0, 2 * np.pi, 100)  # 100 puntos entre 0 y 2*pi
y = np.sin(x)

# Crear la gráfica con color rojo
plt.plot(x, y, color='red')

# Agregar etiquetas y título
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.title("Gráfica de la función seno")

# Mostrar la gráfica
plt.grid(True)
plt.show()
```
